```python
lgr_metrics = pd.DataFrame(columns = ['Features', 'Confusion Matrix', 'Precision', 'Recall', 'F1 Score'])
```

**Feature Scaling**

In logistic regression, we will do feature scaling because we want accurate result of predictions. Here we will only scale the independent variable because dependent variable have only 0 and 1 values. Below is the code for it:

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train_1 = st_x.fit_transform(x_train)
x_test_1 = st_x.transform(x_test)
```

```python
# We create a Logistic Regression object
classifier= LogisticRegression(random_state=0)
# We fit the model using the feature Time
classifier.fit(x_train_1, y_train)
#Predicting the test set result
y_pred= classifier.predict(x_test_1)

#Creating the Confusion matrix, Precision, Recall, F1-Score
cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print('Confusion Matrix:\n', cm)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)
lgr_metrics = lgr_metrics._append({'Features': ['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9',
                                                'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19',
                                                'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount'],
                                   'Confusion Matrix':cm, 'Precision':precision, 'Recall':recall, 'F1 Score':f1}, ignore_index = True)
```
